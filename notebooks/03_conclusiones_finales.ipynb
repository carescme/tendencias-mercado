# Fase 4: Consolidación de Insights y Conclusiones Finales

En este notebook se realiza el cierre del análisis de tendencias de búsqueda para las 20 Inteligencias Artificiales seleccionadas (periodo 2022-2025). 

### Nota sobre el Criterio de Análisis:
Se ha decidido **omitir un modelo predictivo automatizado** (como la Regresión Lineal) para este conjunto de datos. La razón es la **extrema volatilidad** del mercado de IA:
1. **Disrupción constante:** El lanzamiento de un nuevo modelo (ej. Sora o GPT-5) invalida inmediatamente las tendencias históricas.
2. **Naturaleza del Mercado:** En sectores tecnológicos emergentes, el comportamiento pasado no garantiza el futuro, por lo que un análisis comparativo y de ratios de crecimiento aporta mayor valor estratégico que una predicción matemática lineal.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga de datos
df = pd.read_csv('../datos/datos_tendencias_20_IAs.csv')
df['fecha'] = pd.to_datetime(df['fecha'])
```

```python
# Configuración visual
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 7))
```

```python
# Selección de IAs para el resumen visual (Líderes de mercado)
ias_principales = ['ChatGPT', 'Gemini', 'Sora', 'Claude', 'Midjourney']

for ia in ias_principales:
    plt.plot(df['fecha'], df[ia], label=ia, linewidth=2.5, alpha=0.8)

plt.title('Resumen Ejecutivo: Evolución de las IAs con Mayor Impacto', fontsize=16)
plt.xlabel('Periodo Analizado', fontsize=12)
plt.ylabel('Índice de Interés (Google Trends)', fontsize=12)
plt.legend(title='Herramientas IA', loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)

plt.xticks(rotation=45)

# Guardar la imagen final para el reporte
plt.savefig('../reportes/resumen_ejecutivo_final.png')
plt.show()
```

## Conclusiones Generales del Proyecto

Tras el análisis exhaustivo de los datos de Google Trends, se extraen las siguientes conclusiones de mercado:

1. **Hegemonía de ChatGPT:** Mantiene el volumen de búsqueda más alto de forma sostenida. Se ha consolidado como la marca "paraguas" de la IA para el público general.
2. **Explosividad vs. Estabilidad:** * **Sora y Gemini** muestran picos de crecimiento masivos en tiempos récord, lo que indica que el mercado reacciona cada vez más rápido a los anuncios de IA generativa.
   * Herramientas como **GitHub Copilot** o **Stable Diffusion** mantienen líneas más planas; esto no indica falta de éxito, sino una **adopción de nicho profesional** donde el volumen de búsqueda es constante pero no masivo.
3. **Validación de Datos:** Los ratios de crecimiento calculados (81x para ChatGPT, 5x para Sora) demuestran que el interés del mercado no es solo alto, sino que crece de forma exponencial tras cada hito tecnológico.
