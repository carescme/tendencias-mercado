```python
import pandas as pd
from pytrends.request import TrendReq
import time
```

```python
### CONFIGURACION
pytrends = TrendReq(hl='es-ES', tz=360)

# IAs a analizar divididas de 5 en 5 (máximo de la API)
palabras_clave = [
    ['ChatGPT', 'Midjourney', 'GitHub Copilot', 'Gemini', 'Perplexity'],
    ['Claude', 'DALL-E', 'Code Llama', 'Llama', 'Kimi'],
    ['Grok', 'Stable Diffusion', 'Tabnine', 'Mistral AI', 'You.com'],
    ['Microsoft Copilot', 'Runway Gen-2', 'AlphaCode', 'Sora', 'DeepMind']
]
```

```python
###PARAMETROS DE TIEMPO Y LOCALIZACION
# Consulta desde el inicio de 2022 hasta ayer (01/12/2025)
TIME_FRAME = '2022-01-01 2025-12-01' 
PAIS = 'ES' # Foco en España
todos_datos = []
```

```python
### EXTRACCION DE DATOS
for i, grupo in enumerate(palabras_clave):
    print(f"\n- Extrayendo Grupo {i+1}: {', '.join(grupo)}") 

    try:
        #Petición API
        pytrends.build_payload(
            kw_list=grupo,
            cat=0,
            timeframe=TIME_FRAME,
            geo=PAIS
        )

        #Obtener datos de interés a lo largo del tiempo
        df_grupo = pytrends.interest_over_time()
    
        #Limpieza: Eliminar la columna 'isPartial' y renombrar la fecha
        df_grupo = df_grupo.drop(columns=['isPartial'], errors='ignore')
        df_grupo.reset_index(inplace=True)
        df_grupo.rename(columns={'date': 'fecha'}, inplace=True)
    
        todos_datos.append(df_grupo)
        time.sleep(5) #Pausa evitar bloqueo API

    except Exception as e:
        print(f"Error al extraer el Grupo {i+1}: {e}")
        time.sleep(10)
```

```python
### ENSAMBLAJE FINAL
# Usamos pd.merge para unir los DataFrames secuencialmente usando 'fecha' como clave.
df_final = todos_datos[0]
for df_temp in todos_datos[1:]:
    df_final =pd.merge(df_final, df_temp, on='fecha', how='outer')
```

```python
### ALMACENAR DATOS EN CSV
df_final.to_csv('../datos/datos_tendencias_20_IAs.csv', index=False)
print(f"\n Extracción de 20 IAs completada. Datos guardados.")
```
